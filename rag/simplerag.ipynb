```python
from langchain_community.document_loaders import UnstructuredURLLoader

loader = UnstructuredURLLoader(
    urls=["https://lilianweng.github.io/posts/2023-06-23-agent/"]
)

docs = loader.load()
docs
```

```python
##loading a pdf 
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("13b.pdf")
documents = loader.load()
documents
```

### Till now load part is done, lets go with transform part

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=200)
documen=text_splitter.split_documents(documents)
documen[:5]
```

## Now converting the chunks into vectors

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import OpenAIEmbeddings
db = Chroma.from_documents(documen[:10],OpenAIEmbeddings())
```

```python
query = "What the research paper is about?"
result = db.similarity_search(query)
result[0].page_content
```
